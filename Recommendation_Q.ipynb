```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data = pd.read_csv("P:/python/book(RS).csv")
data
```

```python
data = data.drop('Unnamed: 0', axis = 1)
data
```

```python
data1 = data.pivot_table(index = 'User.ID', columns = 'Book.Title', values = 'Book.Rating')
data1
```

```python
data1.fillna(0,inplace=True)
data1
```

```python
data1_dist = 1-pairwise_distances(data1.values, metric = 'cosine')
np.fill_diagonal(data1_dist, 0)
data1_dist1 = pd.DataFrame(data1_dist)
data1_dist1
```

```python
data1_dist1.index=data['User.ID'].unique()
data1_dist1.columns=data['User.ID'].unique()
data1_dist1
```
