```python
import pandas as pd
```

```python
data = pd.read_csv("P:/python/book(RS).csv")
data
```

```python
data = data.drop('Unnamed: 0', axis = 1)
data = data.rename({'Book.Title':'book_title','Book.Rating':'book_rating','User.ID':'userid'}, axis = 1)
data
```

```python
data = data.sort_values('userid').reset_index()
data = data.drop('index', axis = 1)
data
```

```python
data.info()
```

```python
data1 = data.pivot_table(index = 'book_title', columns = 'userid', values = 'book_rating').fillna(0)
data1
```

```python
from scipy.sparse import csr_matrix
data1_matrix = csr_matrix(data1.values)
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(data1_matrix)
```

```python
query_index = 567
distances, indices = model_knn.kneighbors(data1.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(data1.index[query_index]))
    else:
        print('{0}: {1}'.format(i, data1.index[indices.flatten()[i]]))
```
